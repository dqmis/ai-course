```python
# Iterators
from sys import getsizeof
from typing import List

class StarsInGalaxy:
    LIMIT = 100000000

    def __init__(self) -> None:
        self._value = 1
    
    def __iter__(self):
        return self

    def __next__(self) -> int:
        if self._value > self.LIMIT:
            raise StopIteration
        i = self._value
        self._value += 1
        return i

class Cats:
    def __init__(self) -> None:
        self._cats: List[str] = []
        self._cur = 0

    def __iter__(self):
        return self

    def __next__(self) -> str:
        i = self._cur  
        if i >= len(self._cats):
            raise StopIteration
        self._cur += 1
        return self._cats[i]

    def __add__(self, _cat: str):
        self._cats.append(_cat)
        return self
```

```python
stars_in_galaxy = StarsInGalaxy()
print("Size in memory:", getsizeof(stars_in_galaxy))
type(stars_in_galaxy)

print(next(stars_in_galaxy))
print(next(stars_in_galaxy))
print(next(stars_in_galaxy))
print(next(stars_in_galaxy))

#for num in stars_in_galaxy:
    #print(num)
```

```python
class CatsWithReset:
    def __init__(self) -> None:
        self._cats: List[str] = []
        self._cur = 0

    def __iter__(self):
        return self

    def __next__(self) -> str:
        i = self._cur  
        if i >= len(self._cats):
            raise StopIteration
        self._cur += 2
        return self._cats[i]

    def __add__(self, _cat: str):
        self._cats.append(_cat)
        self._cur = 0 
        return self
```

```python
cats = CatsWithReset()
cats += "blue"
cats += "pilkis"
cats += "pukis"
for cat in cats:
    print(cat)
cats += "melyne"
print(">"*10)
for cat in cats:
    print(cat)
```

```python
import requests, io, cv2

from typing import Optional
import numpy as np
from PIL import Image


def url_to_rgb_values(url: str) -> Optional[np.ndarray]:
    try:
        response = requests.get(url)
        bytes_im = io.BytesIO(response.content)
        cv_im = np.array(Image.open(bytes_im))
        return cv_im
    except Exception:
        return None
```

```python
class ImagesDataset:
    def __init__(self, image_urls: List[str]) -> None:
        self._data = image_urls
        self._cur = 0
        self._failed_data: List[str] = []

    def __iter__(self):
        return self

    @property
    def failed_data(self) -> List[str]:
        return self._failed_data

    def __next__(self) -> Optional[np.ndarray]:
        i = self._cur
        if i >= len(self._data):
            raise StopIteration
        self._cur += 1
        rgb_image = url_to_rgb_values(self._data[i]) 
        if rgb_image is not None:
            return rgb_image
        self._failed_data.append(self._data[i])
        return None

list_of_rgbs = [url_to_rgb_values(url) for url in list_of_images]
images_dataset = ImagesDataset(list_of_images)

print("Size of list:", getsizeof(list_of_rgbs))
print("Size of dataset:", getsizeof(images_dataset))

for image in list_of_rgbs:
    plt.imshow(image)
    plt.show()

for image in images_dataset:
    plt.imshow(image)
    plt.show()
```

```python
import time

def timer(fn):
    def track_time(*args, **kwargs):
        start_time = time.time()
        res = fn(*args, **kwargs)
        execution_time = f"Execution time {time.time() - start_time}"
        return res, execution_time
    return track_time

def logger_fun(fn):
    def log_fun(*args, **kwargs):
        res, time = fn(*args, **kwargs)
        logging.info(f"Running a function {fn.__name__} with args {args} and {kwargs}. Run time is {time}")
        return res
    return log_fun

@logger_fun
@timer
def hello(name) -> None:
    print(f"hello! it's me - {name}")

@logger_fun
@timer
def add(a: int, b: int, c: int, d: int) -> int:
    return a + b + c + d   

hello("Dominykas")
add(1, 2, 3, 4)
```
