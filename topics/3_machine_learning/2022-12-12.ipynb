```python
import pandas as pd
from sklearn import tree
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target


clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
df = pd.read_csv("./example_class.csv")
df
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(df[["Loves Winter", "Loves Disney", "Age"]].values, df["Likes Frozen"].values)
```

```python
tree.plot_tree(clf, feature_names=["Loves Winter", "Loves Disney", "Age"])
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, y)
tree.plot_tree(clf)
```

```python
clf = tree.DecisionTreeClassifier(min_samples_leaf=10)
clf = clf.fit(X, y)
tree.plot_tree(clf)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y)
```

```python
clf = tree.DecisionTreeClassifier(min_samples_leaf=10)
clf = clf.fit(x_train, y_train)
print(classification_report(y_test, clf.predict(x_test)))
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(x_train, y_train)
print(classification_report(y_test, clf.predict(x_test)))
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
print(classification_report(y_test, clf.predict(x_test)))
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
print(f1_score(y_test, clf.predict(x_test), average="weighted"))
```

```python
clf = LogisticRegression()
clf = clf.fit(x_train, y_train)
print(f1_score(y_test, clf.predict(x_test), average="weighted"))
```

```python
from sklearn import tree
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

print(X)
print(y)
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X, y)
tree.plot_tree(clf)
```

```python
import matplotlib.pyplot as plt

clf = tree.DecisionTreeRegressor(max_depth=3, min_samples_leaf=10)
clf = clf.fit(X, y)

plt.figure(figsize=(10, 20), dpi=300)
tree.plot_tree(clf)
plt.show()
```

```python
df = pd.read_csv("./example_class.csv")

df["Rating"] = [30, 37, 87, 92, 97, 43, 21]
df
```
